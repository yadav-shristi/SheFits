```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("women_clothing_size_100.csv")
df.head()
```

```python
df.shape
df.info()
df.isnull().sum()
df['Size'].value_counts()
```

```python
plt.figure()
df['Size'].value_counts().plot(kind='bar')
plt.title("Clothing Size Distribution")
plt.xlabel("Size")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
sns.boxplot(x='Size', y='Bust', data=df)
plt.title("Bust vs Clothing Size")
plt.show()
```

```python
plt.figure()
sns.boxplot(x='Size', y='Waist', data=df)
plt.title("Waist vs Clothing Size")
plt.show()
```

```python
plt.figure()
sns.boxplot(x='Size', y='Hip', data=df)
plt.title("Hip vs Clothing Size")
plt.show()
```

```python
plt.figure()
sns.heatmap(df.drop('Size', axis=1).corr(), annot=True)
plt.title("Feature Correlation Heatmap")
plt.show()
```

EDA Summary
- The dataset is clean with no missing values.
- All clothing size classes are evenly distributed.
- Bust, waist, and hip measurements strongly influence clothing size.
- The data is suitable for building a classification model.
